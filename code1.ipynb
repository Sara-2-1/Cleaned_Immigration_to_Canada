```python
import pandas as pd 
df_can=pd.read_csv("Cleaned_Immigration_to_Canada(1).csv")
df_can
```

```python
# 1- Use the dataset Cleaned_Immigration_to_Canada.csv to display the immigration distribution for Greece,
# Albania, and Bulgaria for years 1980 - 2013?
```

```python
df_can.columns = list(map(str, df_can.columns))
years = list(map(str, range(1980,2014)))
years
```

```python
df_can.index.name=None
```

```python
df_can.set_index('Country', inplace = True)
```

```python
df_ct = df_can.loc[['Greece','Albania','Bulgaria', ] , years]
df_ct.head()
```

```python
# 2- Use an overlapping plot with 15 bins and a transparency value of 0.35?
```

```python
df_ct = df_can.loc[['Greece', 'Albania', 'Bulgaria'], years]
df_ct = df_ct.transpose() 
# un-stacked histogram
df_ct.plot(kind ='hist',
            figsize=(10, 6),
            bins=15,
            alpha=0.35,
            color=['red', 'darkslateblue', 'mediumseagreen']
            )

plt.title('Histogram of Immigration from Greece, Albania,  and Bulgaria from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```
